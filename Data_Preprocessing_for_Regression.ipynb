```python
### Merge and Save Y - Train, Valid, and Test

Y_TFR_Train = np.load("Y_TFR_Train.npy").reshape(-1, 1)
Y_ER_Train = np.load("Y_ER_Train.npy").reshape(-1, 1)
Y_MR_Train = np.load("Y_MR_Train.npy").reshape(-1, 1)

Y_TFR_Valid = np.load("Y_TFR_Valid.npy").reshape(-1, 1)
Y_ER_Valid = np.load("Y_ER_Valid.npy").reshape(-1, 1)
Y_MR_Valid = np.load("Y_MR_Valid.npy").reshape(-1, 1)

Y_TFR_Test = np.load("Y_TFR_Test.npy").reshape(-1, 1)
Y_ER_Test = np.load("Y_ER_Test.npy").reshape(-1, 1)
Y_MR_Test = np.load("Y_MR_Test.npy").reshape(-1, 1)

Y_Train = np.concatenate([Y_TFR_Train, Y_ER_Train, Y_MR_Train], axis=1)
Y_Valid = np.concatenate([Y_TFR_Valid, Y_ER_Valid, Y_MR_Valid], axis=1)
Y_Test = np.concatenate([Y_TFR_Test, Y_ER_Test, Y_MR_Test], axis=1)

np.save("Y_Train.npy", Y_Train)
np.save("Y_Valid.npy", Y_Valid)
np.save("Y_Test.npy", Y_Test)
```

```python
### Merge CAE_Encoder_Output_Train and Y_Train

X_Train = np.load("CAE_Encoder_Output_Train.npy")
Y_Train = np.load("Y_Train.npy")

XY_Train = np.concatenate([X_Train, Y_Train], axis=1)

XY_Train_df = pd.DataFrame(XY_Train)

XY_Train_df
```

```python
### Shuffle XY_Train in Row Axis

np.random.seed(42)
np.random.shuffle(XY_Train)

XY_Train_df = pd.DataFrame(XY_Train)

XY_Train_df
```

```python
### Seperate XY_Train into X_Train and Y_Train

CAE_Encoder_Output_Train_Shuffled = XY_Train[:, :34]
Y_Train_Shuffled = XY_Train[:, 34:]

np.save("CAE_Encoder_Output_Train_Shuffled.npy", CAE_Encoder_Output_Train_Shuffled)
np.save("Y_Train_Shuffled.npy", Y_Train_Shuffled)
```
